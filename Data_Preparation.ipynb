```python

# Data Integration
import pandas as pd

# Baca file CSV
df_original = pd.read_csv("Data-Penjualan (Collection Data).csv")

# Cek dulu nama kolom yang tersedia
print("Daftar kolom di dataset:")
print(df_original.columns.tolist())

# Salin dataframe
df_integrasi = df_original.copy()

# Ubah kolom tanggal ke format datetime (hanya jika ada di dataset)
date_columns = ['Tgl Order', 'Tgl DO', 'Tgl FP']
for col in date_columns:
    if col in df_integrasi.columns:
        df_integrasi[col] = pd.to_datetime(df_integrasi[col], errors='coerce')

# Bersihkan dan ubah kolom uang ke format angka
currency_columns = ['Jumlah Bruto (Rp)', 'Total Discount (Rp)', 'Jumlah Netto (Rp)', 'Retur (Rp)', 'Total (Rp)']
for col in currency_columns:
    if col in df_integrasi.columns:
        df_integrasi[col] = (
            df_integrasi[col]
            .astype(str)
            .str.replace(r'[^\d.-]', '', regex=True)
        )
        df_integrasi[col] = pd.to_numeric(df_integrasi[col], errors='coerce')

# Isi data kosong (NaN)
for col in currency_columns:
    if col in df_integrasi.columns:
        df_integrasi[col] = df_integrasi[col].fillna(0)

text_cols = ['Sales', 'Pelanggan', 'Drive', 'Penagih', 'Status']
for col in text_cols:
    if col in df_integrasi.columns:
        df_integrasi[col] = df_integrasi[col].fillna('Unknown')

# Simpan ke CSV
df_integrasi.to_csv("data_integrasi.csv", index=False)
print("File 'data_integrasi.csv' berhasil disimpan.")
```

```python
import pandas as pd

hasil_analisis = {}

# Top 10 Pelanggan
try:
    top_pelanggan = df_integrasi.groupby('Pelanggan')['Jumlah Netto (Rp)'].sum().nlargest(10)
    hasil_analisis["Top 10 Pelanggan Berdasarkan Netto (Rp)"] = top_pelanggan
except KeyError:
    print("⚠️ Kolom 'Pelanggan' atau 'Jumlah Netto (Rp)' tidak ada.")

# Top 10 Sales
try:
    top_sales = df_integrasi.groupby('Sales')['Jumlah Netto (Rp)'].sum().nlargest(10)
    hasil_analisis["Top 10 Sales Berdasarkan Penjualan Netto (Rp)"] = top_sales
except KeyError:
    print("Kolom 'Sales' atau 'Jumlah Netto (Rp)' tidak ada.")

# Simpan ke file TXT
with open("data_analisis.txt", "w", encoding="utf-8") as f:
    for judul, data in hasil_analisis.items():
        f.write(f"{judul}\n")
        f.write(data.to_string())
        f.write("\n\n")

print("Hasil analisis berhasil disimpan ke 'data_analisis.txt'")
```

```python
# Tahap 2 (Bagian 3): Validasi Data

# Salin dataframe
df_validasi = df_integrasi.copy()

# Validasi Duplikat
jumlah_duplikat = df_validasi.duplicated().sum()
df_validasi.drop_duplicates(inplace=True)
jumlah_setelah_dihapus = len(df_validasi)

# Validasi Data Kosong
jumlah_kosong = df_validasi.isnull().sum().sum()

# Simpan ringkasan ke TXT
with open("data_validasi.txt", "w", encoding="utf-8") as f:
    f.write("Validasi Data\n")
    f.write(f"Jumlah baris duplikat: {jumlah_duplikat}\n")
    f.write(f"Jumlah baris setelah hapus duplikat: {jumlah_setelah_dihapus}\n")
    f.write(f"Jumlah baris yang datanya tidak terisi: {jumlah_kosong}\n")

# Simpan data validasi ke CSV juga
df_validasi.to_csv('data_validasi.csv', index=False)

print("File 'data_validasi.csv' dan 'data_validasi.txt' berhasil disimpan.")
```
